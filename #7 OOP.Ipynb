# Object-Oriented Programming (OOP)

## 1. Definition
**Object-Oriented Programming (OOP)** is a programming paradigm centered around the concept of **"objects"** rather than functions or logic. It allows developers to structure software by grouping related data (attributes) and behaviors (methods) into reusable blueprints called **classes**.

*   **Class (Blueprint):** The template or design for creating objects.
*   **Object (Instance):** The actual real-world entity created from the class blueprint.

---

## 2. Analogies to Understand OOP

To better grasp how classes and objects relate, consider these common real-world analogies:

| Concept / Blueprint (Class) | Real-world Instance (Object) | Description |
| :--- | :--- | :--- |
| **House Blueprint** | Physical Houses | One architectural blueprint can be used to construct multiple unique houses (different paint colors, owners, locations). |
| **Car Design / Factory Spec** | Individual Cars | A manufacturer's design template is used to build thousands of physical cars, each with its own license plate and mileage. |
| **Cookie Cutter** | Individual Cookies | A single cookie cutter shape (class) can create dozens of separate cookies (objects), which can then be decorated individually. |

```python


# New Blueprint

class Person:
    def __init__(self,name,age,gender="Male"):
        self.name = name
        self.age = age
        self.gender= gender
    def tellMyName(self):
        print(f"My name is {self.name}")
    def tellMyAge(self):
        print(f"My age is {self.age}")

```

```python


salesPerson= Person("Ali", 30,"Male")

salesPerson.tellMyAge()
salesPerson.tellMyName()

```

```python
# Make an Class of an LED - 
# size, weight, ai_enabled

class LED:
    def __init__(self,size,weight,ai_enabled):
        self.size = size
        self.weight = weight
        self.ai_enabled= ai_enabled
    def get_size(self):
        return self.size
    def get_weight(self):
        return self.weight
    def is_ai_enabled(self):
        return self.ai_enabled
    def get_led_properties(self):
        return {
            "size": self.size,
            "weight": self.weight,
            "ai_enabled": self.ai_enabled
        }
led1=LED(5,0.2,True)
print(led1.get_size())
print(led1.get_led_properties())
```

```python
class Car:
    def __init__(self, make, model, year,ownerInfo):
        self.make = make
        self.model = model
        self.year = year
        self.fuel=0
        self.mileage=0
        self.owner=ownerInfo
    def fuelUp(self, liters):
        self.fuel += liters
        print(f"Filling up {liters} liters of fuel.")

    def getCurrentFuel(self):
        return self.fuel
    
    def getOwnwerInfo(self):
        return self.owner

    def getCurrentMileage(self):
        return self.mileage

    def drive(self, distance):
        if distance * 0.1 > self.fuel:
            print("Not enough fuel to drive the distance.")
            return
        self.mileage += distance
        self.fuel -= distance * 0.1  # Assume fuel consumption is 0.1 liters/km -- 10 km per liter
        print(f"Driving {distance} km.")
```

# Inheritance

```python
class PARENT:
    def showMessage(self):
        print("This is a message from the Parent class.")

class Child(PARENT):
    def printMessage(self):
        print("This is a message from the Child class.")

child_instance = Child()
child_instance.showMessage()  # Inherited from Parent class
child_instance.printMessage()  # Method from Child class
```

```python
class Dog:
    def speak(self):
        return "Woof!"

class Cat(Dog):
    def speak(self):
        return "Meow!"

cat=Cat()
print(cat.speak())
```

```python
class Vehicle:
    def check_fuel(self, side=None):
        if side is None:
            return "Checking fuel level..."
        return f"Checking fuel level on the {side} side..."

My_Car = Vehicle()

print(My_Car.check_fuel())        # No argument
print(My_Car.check_fuel("left"))  # With argument
```
